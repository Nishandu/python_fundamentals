````python
import random

def guess_the_number():
    # 1. The computer selects a random number between 1 and 10
    secret_number = random.randint(1, 10)
    attempts = 3
    
    print("I'm thinking of a number between 1 and 10. You have 3 chances!")

    # 2. The user gets 3 chances to guess the number
    for i in range(attempts):
        try:
            guess = int(input(f"Attempt {i+1}: Enter your guess: "))
            
            # 3. Provide feedback for each guess
            if guess == secret_number:
                print("You guessed it right!")
                return  # 4. Exit the loop/function if correct
            elif guess < secret_number:
                print("Too low.")
            else:
                print("Too high.")
        
        except ValueError:
            print("Please enter a valid whole number.")

    # 5. If they use all attempts
    print(f"Sorry, the number was {secret_number}.")

# Run the game
guess_the_number()
````

````python
def full_pyramid(n):
    for i in range(1, n + 1):
        # Print leading spaces
        for j in range(n - i):
            print(" ", end="")
        
        # Print asterisks for the current row
        for k in range(1, 2*i):
            print("*", end="")
        print()
   
full_pyramid(5)
````

### 16. Create a palindrome checker function.

````python
def is_palindrome_fast(text):
    left = 0
    right = len(text) - 1
    
    while left < right:
        if text[left].lower() != text[right].lower():
            return False
        left += 1
        right -= 1
    return True

print(is_palindrome_fast("maam"))
````

### 20. Create a simple password strength checker.

````python
def check_password_strength(password):
    # 1. Define the criteria
    length_ok = len(password) >= 8
    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_special = any(not char.isalnum() for char in password)

    # 2. Count how many criteria are met
    score = sum([length_ok, has_upper, has_lower, has_digit, has_special])

    # 3. Determine the strength label
    if score <= 2:
        return "Weak"
    elif score <= 4:
        return "Medium"
    else:
        return "Strong "

# --- Test the function ---
pwd = input("Enter a password to check: ")
print(f"Strength: {check_password_strength(pwd)}")
````


**Question:** Create a closure that generates multiplier functions.



````python
def make_multiplier(factor):
    # This is the enclosing function
    
    def multiply(number):
        # This nested function "closes over" the 'factor' variable
        return number * factor
    
    return multiply  # Returns the function itself, not the result

# Create specific multiplier functions
double = make_multiplier(2)
triple = make_multiplier(3)

# Use the generated functions
print(double(10)) # Output: 20
print(triple(10)) # Output: 30
````

### Assignment 1: Understanding OOP vs Procedural Programming
#### Theory Question
Explain the key differences between procedural and object-oriented programming with examples. Include at least three differences and real-world analogies.

### Key Differences
1. Procedural programming focuses on the sequence of functions and logic, while OOP focuses on organizing data into objects.

2. Procedural follows a top-down approach starting from the main goal, whereas OOP follows a bottom-up approach by building reusable components.

3. Procedural programs typically use global data accessible to all functions, while OOP uses encapsulation to keep data private within a class.10

### Real-World Analogies
1. Following a cooking recipe step-by-step is procedural, while using a microwave as a self-contained unit with its own settings is object-oriented.

2. A factory assembly line moving through specific stations is procedural, while a car with its own internal parts and behaviors is an object.

3. Writing a handwritten letter from start to finish is procedural, while using a smartphone with various apps and stored data is object-oriented.

````python
class ShoppingCart:
    def __init__(self):
        self.cart = {}  #dictionary that contains item name as key and quantity as value
    
    def add_item(self, item, quantity=1):
        if item in self.cart:
            self.cart[item] += quantity #increasing the quantity in dictionary by quantity if the key already exist in dictionary
        else:
            self.cart[item] = quantity #if not in dictionary , creating the key value pair by item and quantity
    
    def remove_item(self, item, quantity=1):
        if item in self.cart:
            items_to_remove =  self.cart[item] - quantity #how many items to remove 
            self.cart[item] = max(0,items_to_remove ) #max to modify the quantity value in cart dictionary for the item
            if self.cart[item] == 0: #if the items value is 0 just deleting it
                del self.cart[item]

    def show_all_items(self):
        print(f'The items in the cart are: \n ')
        for i,item in enumerate(self.cart.items()): #assigning index to each keys while we iterate to print item number
            item_name, quantity = item
            print(f"Item {i} : {quantity} {item_name}")
    

cart = ShoppingCart()
cart.add_item("Apple", 4 )
cart.add_item("Banana", 6)


cart.show_all_items()
````

## Static Methods, Class Methods & Dunder Methods

### Understanding Method Types:

1. **Static Methods**
   - Don't have access to `cls` or `self`
   - Decorated with `@staticmethod`
   - Used for utility functions related to the class
   - Can't modify class or instance state

2. **Class Methods**
   - Have access to `cls` (the class)
   - Decorated with `@classmethod`
   - Can modify class state
   - Can create class instances

3. **Dunder Methods**
   - Special methods with double underscores
   - Automatically called in specific situations
   - Customize class behavior

### Exercise 3: Book Class (Dunder Methods)

#### Key Dunder Methods:
1. `__init__`: Constructor
2. `__str__`: String representation
3. `__eq__`: Equality comparison


````python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    # String representation (for print)
    def __str__(self):
        return f"'{self.title}' by {self.author}, Price: {self.price}"

    # Equality comparison (==)
    def __eq__(self, other):
        if isinstance(other, Book):
            return self.title == other.title and self.author == other.author
        return False
````
